```python
# Revenue generated per night

bookings["Revenue_Per_Night"] = (
    bookings["Total_Revenue"]
    /
    bookings["Length_of_Stay"]
)
```

```python
# Create column Booking month

bookings["Booking_Month"] = (
    pd.to_datetime(bookings["Booking_Date"])
    .dt.month_name()
)
```

```python
# Create column Check-in month

bookings["Check_In_Month"] = (
    pd.to_datetime(bookings["Check_In_Date"])
    .dt.month_name()
)
```

```python
# Create column Stay category

def stay_category(days):

    if days <= 3:
        return "Short Stay"

    elif days <= 7:
        return "Medium Stay"

    else:
        return "Long Stay"


bookings["Stay_Category"] = (
    bookings["Length_of_Stay"]
    .apply(stay_category)
)
```

```python
bookings.head()
```

```python
# Create column Month

hotel_kpi["Month"] = (
    pd.to_datetime(hotel_kpi["Date"])
    .dt.month_name()
)
```

```python
# Create column Year

hotel_kpi["Year"] = (
    pd.to_datetime(hotel_kpi["Date"])
    .dt.year
)
```

```python
# Validate Revenue per Available Room

hotel_kpi["Calculated_RevPAR"] = (
    hotel_kpi["Room_Revenue"]
    /
    hotel_kpi["Available_Rooms"]
)
```

```python
hotel_kpi.head()
```

```python
# Create column Pricing Index

competitor["Pricing_Index"] = (
    competitor["Hilton_ADR"]
    /
    competitor["Market_Average_ADR"]
)
```

```python
competitor.head()
```

```python
# Create column Review Month

reviews["Review_Month"] = (
    pd.to_datetime(reviews["Review_Date"])
    .dt.month_name()
)
```

```python
# Create column Rating Category

def rating_category(rating):

    if rating >=4:
        return "Satisfied"

    elif rating ==3:
        return "Neutral"

    else:
        return "Dissatisfied"


reviews["Rating_Category"] = (
    reviews["Rating"]
    .apply(rating_category)
)
```

```python
reviews.head()
```

```python
# Create column Events Month

events["Event_Month"] = (
    pd.to_datetime(events["Date"])
    .dt.month_name()
)
```
